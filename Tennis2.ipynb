### We scraped all the information needed from the ATP website in order to cluster the players and the court type later on. For each tournament, we scraped the name of the tournament, whether it is indoor/outdoor, and the surface of the court. All the information is stored in the CSV file called tennisRatings.csv. Before the final report, we plan on clustering in order to predict the best court type to play on###

```python
def getScores(html, file):
    name = []
    cn = []
    tp = []
    match_soup = BeautifulSoup(html, "lxml")
  
    # first loop through all 67 tournaments and get the names
    for i in range (0,67):
        match = match_soup.find_all('span', class_ = 'tourney-title')[i].find(text=True)
        match = ' '.join(match.split())
        name.append(match)
        
        
    for i in range (0,201):#201
        details = match_soup.find_all('div', class_ = 'item-details')[i].find(text=True)
        details = ' '.join(details.split())
        print(details)
        cn.append(details)

        value = match_soup.find_all('div', class_ = 'item-details')[i].find('span', class_ = 'item-value').find(text=True)
        value = ' '.join(value.split())
        tp.append(value)
        print(value)
        
    k = 1
    i = 0
    for tournament in name:
        match = name[i]
        courtLoc = cn[k]
        typeCourt = tp[k]
        k = k + 3
        i += 1
        try:
            file.write(match + "," + courtLoc + "," + typeCourt)
            file.write("\n")
        except:
            continue

    
```
